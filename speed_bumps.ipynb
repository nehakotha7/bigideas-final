# Sub-metric #1 : Speed Bumps (Rand)

For the first sub-metric, I chose to take a look at data gathered by the City of Pittsburgh about speed bumps.  While there was a lot of data from the .csv file, I manipulated it to highlight the information I cared about in my analysis.

```python
import pandas as pd

data = pd.read_csv("speed_bumps.csv", sep=',')
print(data['neighborhood'].unique())
```

Above is a list of all of the neighhorhoods with speed bumps.  Notice that there are many Pittsburgh neighborhoods not listed.  Those neighborhoods are the ones the other sub-metrics may be able to gather more insight about to figure out which is the best.  The ones I picked out are now lower on our list of the "best" to speed in.  To truly rank them though, I counted how many speed bumps were in the neighborhoods in that list.

```python
items = dict()

for index,row in data.iterrows():
    if row['neighborhood'] in items:
        items[row['neighborhood']] += 1
    else:
        items[row['neighborhood']] = 1
print(items)
```

Let's view this data in a clearer format:

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

neighborhoods = list(items.keys())
num_bumps = list(items.values())

plt.bar(neighborhoods, num_bumps, color="green")
plt.title("Number of Speed Bumps in Pittsburgh Neighborhoods")
plt.xlabel("Neighborhoods")
plt.xticks(rotation = 90)
plt.ylabel("Number of Speed Bumps")
plt.show()
```

Much better!  From this bar graph, we can see that Squirrel Hill South has the most speed bumps by far!  Definitely not the place to speed in.  Central Oakland, Stanton Heights, Overbrook, and Highland Park have the least so there's still a shot they'd be okay to speed in... depends on the other sub-metrics analysis'.  Everything in between is likely not great to speed in.  To reiterate, this dataset is a good introduction to what we can expect the result NOT to be.
